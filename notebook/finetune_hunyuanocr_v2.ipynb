```python
from transformers import AutoProcessor, HunYuanVLForConditionalGeneration
from PIL import Image
import torch
from peft import PeftModel  # Cần thiết để load LoRA

def scale_image_limit(image: Image.Image, max_pixels: int = 1536*1536) -> Image.Image:
    width, height = image.size
    current_pixels = width * height
    if current_pixels <= max_pixels:
        return image
    import math
    ratio = math.sqrt(max_pixels / current_pixels)
    return image.resize((int(width * ratio), int(height * ratio)), Image.Resampling.LANCZOS)

base_model_path = "tencent/HunyuanOCR" 
lora_path = "jaqja/HunYuanOCR-VNCLURv2"

prompt_1 = "Extract all information from the main body of a Vietnam Certificate of Land Use Rights image (Giấy chứng nhận quyền sử dụng đất của Việt Nam). Present any tables in Markdown format. Replace non-text elements such as land plot diagrams, seals, signatures, or images with [short descriptions in Vietnamese enclosed in brackets]. Ensure the parsing is accurate, preserves the original meaning, and is organized clearly following the natural reading order of the document."
# prompt_2 = "Extract all information from the main body of the Vietnam legal document image. Replace the seal and signature with [short descriptions in Vietnamese enclosed in brackets] such as [Con dấu], [Chữ ký] or [Con dấu và chữ ký]. The parsing should be organized according to the reading order."
# prompt_3 = "Extract the handwritten Vietnamese text in the image."
# prompt_4 = "This is one of the image pages in Vietnam's regulations on land use right certificates. Extract all main content exactly, tables should be expressed in Markdown format, non-text elements such as diagrams, seals, and signatures should be replaced with [short descriptions in Vietnamese enclosed in brackets]. The parsing should be organized according to the reading order."

processor = AutoProcessor.from_pretrained(lora_path, use_fast=False)

model = HunYuanVLForConditionalGeneration.from_pretrained(
    base_model_path,
    attn_implementation="eager",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

model = PeftModel.from_pretrained(model, lora_path, subfolder="checkpoint-594")
model.eval() 
```

```python
import os
from pathlib import Path
from tqdm import tqdm

input_folder = "test"
output_folder = "result"
os.makedirs(output_folder, exist_ok=True)

image_extensions = (".png", ".jpg", ".jpeg", ".PNG", ".JPG", ".JPEG")
image_files = [f for f in os.listdir(input_folder) if f.endswith(image_extensions)]

print(f"Tìm thấy {len(image_files)} ảnh. Bắt đầu chạy inference...")

for file_name in tqdm(image_files):
    img_path = os.path.join(input_folder, file_name)

    try:
        raw_image = Image.open(img_path).convert("RGB")
        image_inputs = scale_image_limit(raw_image)

        messages1 = [
            {"role": "system", "content": ""},
            {
                "role": "user",
                "content": [
                    {"type": "image", "image": img_path},
                    {"type": "text", "text": prompt_1},
                ],
            }
        ]

        text_template = processor.apply_chat_template(messages1, tokenize=False, add_generation_prompt=True)

        inputs = processor(
            text=[text_template],
            images=[image_inputs],
            padding=True,
            return_tensors="pt",
        ).to(model.device)

        with torch.no_grad():
            generated_ids = model.generate(
                **inputs,
                max_new_tokens=5000,
                do_sample=False
            )

        input_ids = inputs.input_ids if "input_ids" in inputs else inputs.inputs
        generated_ids_trimmed = generated_ids[0][len(input_ids[0]):]

        output_text = processor.decode(
            generated_ids_trimmed,
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False
        )

        result_file_name = Path(file_name).stem + ".txt"
        result_path = os.path.join(output_folder, result_file_name)

        with open(result_path, "w", encoding="utf-8") as f:
            f.write(output_text)

    except Exception as e:
        print(f"Lỗi khi xử lý ảnh {file_name}: {e}")

print(f"--- Hoàn thành! Kết quả đã được lưu tại thư mục '{output_folder}' ---")
```
