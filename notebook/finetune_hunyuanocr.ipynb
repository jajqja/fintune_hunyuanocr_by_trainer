```python
from transformers import AutoProcessor
from transformers import HunYuanVLForConditionalGeneration
import torch

model_name_or_path = "HunYuanOCR-VNCLUR"
processor = AutoProcessor.from_pretrained(model_name_or_path, use_fast=False)
model = HunYuanVLForConditionalGeneration.from_pretrained(
    model_name_or_path,
    attn_implementation="eager",
    dtype=torch.bfloat16,
    device_map="auto"
)
```

```python
import os
from pathlib import Path
from tqdm import tqdm
from PIL import Image

def scale_image_limit(image: Image.Image, max_pixels: int = 2000000) -> Image.Image:
    width, height = image.size
    current_pixels = width * height

    if current_pixels <= max_pixels:
        return image

    import math
    ratio = math.sqrt(max_pixels / current_pixels)

    new_width = int(width * ratio)
    new_height = int(height * ratio)

    return image.resize((new_width, new_height), Image.Resampling.LANCZOS)

input_folder = "test"
output_folder = "result"
os.makedirs(output_folder, exist_ok=True)
PROMPT = "Extract all text from main body of the image of the Vietnam Certificate of Land Use Rights. Tables should be expressed in Markdown format. Ensure the parsing follows the logical reading order. Replace the land-plot diagram with [Sơ đồ], seal with [Con dấu] and signature with [Chữ ký]."

image_extensions = (".png", ".jpg", ".jpeg", ".PNG", ".JPG", ".JPEG")
image_files = [f for f in os.listdir(input_folder) if f.endswith(image_extensions)]

print(f"Tìm thấy {len(image_files)} ảnh. Bắt đầu chạy inference...")

for file_name in tqdm(image_files):
    img_path = os.path.join(input_folder, file_name)

    try:
        raw_image = Image.open(img_path).convert("RGB")
        image_inputs = scale_image_limit(raw_image)

        messages1 = [
            {"role": "system", "content": ""},
            {
                "role": "user",
                "content": [
                    {"type": "image", "image": img_path},
                    {"type": "text", "text": PROMPT},
                ],
            }
        ]

        text_template = processor.apply_chat_template(messages1, tokenize=False, add_generation_prompt=True)

        inputs = processor(
            text=[text_template],
            images=[image_inputs],
            padding=True,
            return_tensors="pt",
        ).to(model.device)

        with torch.no_grad():
            generated_ids = model.generate(
                **inputs,
                max_new_tokens=4096,
                do_sample=False
            )

        input_ids = inputs.input_ids if "input_ids" in inputs else inputs.inputs
        generated_ids_trimmed = generated_ids[0][len(input_ids[0]):]

        output_text = processor.decode(
            generated_ids_trimmed,
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False
        )

        result_file_name = Path(file_name).stem + ".txt"
        result_path = os.path.join(output_folder, result_file_name)

        with open(result_path, "w", encoding="utf-8") as f:
            f.write(output_text)

    except Exception as e:
        print(f"Lỗi khi xử lý ảnh {file_name}: {e}")

print(f"--- Hoàn thành! Kết quả đã được lưu tại thư mục '{output_folder}' ---")
```
